```python
missing_df = finalmerge.isnull().sum(axis=0).reset_index()
missing_df.columns = ['column_name', 'missing_count']
missing_df = missing_df.loc[missing_df['missing_count']>0]
missing_df = missing_df.sort_values(by='missing_count')

ind = np.arange(missing_df.shape[0])
width = 0.9
fig, ax = plt.subplots(figsize=(12,18))
rects = ax.barh(ind, missing_df.missing_count.values, color='#34495e')
ax.set_yticks(ind)
ax.set_yticklabels(missing_df.column_name.values, rotation='horizontal')
ax.set_xlabel("Count of missing values")
ax.set_title("Number of missing values in each column")
plt.show()
```

```python
finalmerge.fillna(finalmerge.mean(), inplace=True)
```

```python
def datesplitter(df):
    df['year']=df['transactiondate'].dt.year
    df['month']=df['transactiondate'].dt.month
    df['day']=df['transactiondate'].dt.day
    df.drop(['transactiondate'],inplace=True,axis=1)
    return df
```

```python
finalmerge = datesplitter(finalmerge)
```

```python
cols=finalmerge.columns.tolist()
print(cols)
```

```python
cols=cols[:2]+cols[-3:]+cols[2:-3]
finalmerge=finalmerge[cols]
```

```python
finalmerge.head()
```

```python
X = finalmerge.drop(['logerror', 'parcelid', 'propertycountylandusecode' ,'propertyzoningdesc'],axis=1).values
y = finalmerge['logerror'].values

from sklearn.model_selection import train_test_split

X_train,X_val,y_train,y_val=train_test_split(X,y,test_size=0.25)
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error

XG = XGBRegressor(n_estimators=100,learning_rate=0.05,eval_metric='mae')
XG.fit(X_train,y_train,early_stopping_rounds=5,eval_set=[(X_val,y_val)],verbose=True)
pred = XG.predict(X_val)
print(mean_absolute_error(y_val,pred))
```

```python
sample = pd.read_csv("sample_submission.csv")
```

```python
sample.head()
```

```python
sample.shape
```

```python
test_df = pd.merge(sample[['ParcelId']],prop1.rename(columns = {'parcelid':'ParcelId'}), how='left', on='ParcelId')
```

```python
to_remove=[]
for each in test_df.columns:
    if each not in finalmerge.drop(['logerror', 'propertycountylandusecode' ,'propertyzoningdesc'],axis=1).rename(columns = {'parcelid':'ParcelId'}).columns:
        to_remove.append(each)
```

```python
to_remove
```

```python
test_df=test_df.drop(to_remove,axis=1)
```

```python
test_df.mean()
```

```python
test_df.fillna(test_df.mean(), inplace=True)
```

```python
test_df.head()
```

```python
predict=pd.DataFrame({
    'ParcelId':test_df['ParcelId']
})
```

```python
predict.head()
```

```python
test_dates = {
    '201610': pd.Timestamp('2016-09-30'),
    '201611': pd.Timestamp('2016-10-31'),
    '201612': pd.Timestamp('2016-11-30'),
    '201710': pd.Timestamp('2017-09-30'),
    '201711': pd.Timestamp('2017-10-31'),
    '201712': pd.Timestamp('2017-11-30')
}
```

```python
for each in test_dates:
    test_df['transactiondate']=test_dates[each]
    
    test_df=datesplitter(test_df)
    
    cols=test_df.columns.tolist()
    cols=cols[-3:]+cols[:-3]
    test_df=test_df[cols]
    
    X_test=test_df.drop(['ParcelId'],axis=1).values
    ans=XG.predict(X_test)
    
    predict[each]=ans
    
    del test_df['year']
    del test_df['month']
    del test_df['day']
```
